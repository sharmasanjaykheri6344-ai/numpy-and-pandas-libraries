Q.1  Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the
dataset information to understand its structure?

....>> Here’s a simple Python example to load a retail sales dataset into a Pandas DataFrame and inspect it:

import pandas as pd

# Load dataset (replace 'retail_sales.csv' with your file path)
df = pd.read_csv('retail_sales.csv')

df.head() → Shows the first 5 rows
df.info() → Provides structure details like:
Column names
Data types
Non-null values

Q.2 . Inspect the dataset and identify missing values in each column. Display the total number of missing
values present in the dataset.

...>> You can inspect missing values in your dataset using Pandas with the following code:

import pandas as pd

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Check missing values in each column
print("Missing values in each column:")

Explanation:

df.isnull().sum().sum() → Gives total missing values in the entire dataset

Q.3 Replace missing values in the Quantity and Price columns with their respective mean values using
Pandas.

...>> You can replace missing values in the Quantity and Price columns with their respective mean values using Pandas like this:

import pandas as pd

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Replace missing values with mean
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price'].fillna(df['Price'].mean(), inplace=True)

# Verify changes
print(df[['Quantity', 'Price']].isnull().sum())
Explanation:
df['Quantity'].mean() → Calculates the average of the Quantity column
fillna() → Replaces missing (NaN) values with the mean
Same logic applies to the Price column
Alternative (without inplace):
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

Q.5 Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.

.....>> You can create a new Revenue column using NumPy like this:

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Create Revenue column
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

Explanation:
np.multiply(df['Quantity'], df['Price']) → Multiplies Quantity and Price element-wise
The result is stored in a new column called Revenue
Alternative (simpler way without NumPy):
df['Revenue'] = df['Quantity'] * df['Price']

Both methods give the same result ✅.

Q.8 Based on the revenue calculated for each product category, identify the top 3 categories with the
highest revenue and the bottom 3 categories with the lowest revenue.

import pandas as pd

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Group by Product Category
category_revenue = df.groupby('Product Category')['Revenue'].sum().reset_index()

top_3 = category_revenue.sort_values(by='Revenue', ascending=False).head(3)

# Bottom 3 categories (lowest revenue)

print("Top 3 Categories by Revenue:")

print(bottom_3)

Q.9  Group the dataset by Region and calculate the total revenue generated in each region. Identify the
region with the highest revenue and the region with the lowest revenue.

...>> You can group the dataset by Region, calculate total revenue, and identify the highest and lowest regions like this:

import pandas as pd

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Group by Region and calculate total revenue
region_revenue = df.groupby('Region')['Revenue'].sum().reset_index()

# Find region with highest revenue
highest_region = region_revenue.loc[region_revenue['Revenue'].idxmax()]

# Find region with lowest revenue
lowest_region = region_revenue.loc[region_revenue['Revenue'].idxmin()]

# Display results
print("Revenue by Region:")
print(region_revenue)

print("\nRegion with Highest Revenue:")
print(highest_region)

print("\nRegion with Lowest Revenue:")
print(lowest_region)
Explanation:
groupby('Region') → Groups data by region
sum() → Calculates total revenue per region
idxmax() → Finds index of maximum revenue
idxmin() → Finds index of minimum revenue
Example Output:

East    30000
West    15000

Highest Revenue Region:
North   40000

Q.10 Extract the month from the Date column using Pandas and group the dataset by month to calculate the
total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
using NumPy.

...>> You can extract the month, calculate monthly revenue, and compute statistics using NumPy like this:

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('retail_sales.csv')

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract month
df['Month'] = df['Date'].dt.month

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Group by Month and calculate total revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()

# Calculate statistics using NumPy
mean_revenue = np.mean(df['Revenue'])
median_revenue = np.median(df['Revenue'])
std_revenue = np.std(df['Revenue'])

# Display results
print("Monthly Revenue:")
print(monthly_revenue)

print("\nRevenue Statistics:")
print("Mean:", mean_revenue)
print("Median:", median_revenue)
print("Standard Deviation:", std_revenue)
Explanation:
pd.to_datetime() → Converts Date column to proper datetime format
.dt.month → Extracts month from date
groupby('Month') → Aggregates revenue month-wise
np.mean(), np.median(), np.std() → Calculate statistical measures
Example Output:
Month   Revenue
1       20000
2       25000
3       18000

Mean: 220.5
Median: 200.0
Standard Deviation: 50.3.
